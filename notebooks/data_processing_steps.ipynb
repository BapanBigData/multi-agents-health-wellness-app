```python
base_columns = [
    "NPI",
    "Entity Type Code",
    "Provider Organization Name (Legal Business Name)",
    "Provider Last Name (Legal Name)",
    "Provider First Name",
    "Provider Business Practice Location Address City Name",
    "Provider Business Practice Location Address State Name",
    "Provider Business Practice Location Address Postal Code",
    "Provider First Line Business Practice Location Address",
    "Provider Business Practice Location Address Telephone Number",
]

# Step 2: Flatten taxonomy fields into two new features
def extract_taxonomy_fields(row):
    taxonomy_codes = []
    primary_code = None
    for i in range(1, 16):
        code = row.get(f"Healthcare Provider Taxonomy Code_{i}")
        primary_switch = row.get(f"Healthcare Provider Primary Taxonomy Switch_{i}")
        if pd.notna(code):
            taxonomy_codes.append(code)
            if primary_switch == "Y":
                primary_code = code
    return pd.Series({
        "Taxonomy Codes": taxonomy_codes,
        "Primary Taxonomy Code": primary_code
    })
```

```python
# Step 3: Apply transformation and merge with base columns
taxonomy_df = df.apply(extract_taxonomy_fields, axis=1)
df_filtered = pd.concat([df[base_columns], taxonomy_df], axis=1)
```

```python
df_filtered["Provider Business Practice Location Address Telephone Number"] = df_filtered["Provider Business Practice Location Address Telephone Number"].apply(lambda x: x.split(".")[0])
```

```python
# Load NUCC taxonomy reference data
taxonomy_df = pd.read_csv("../data/nucc_taxonomy_250.csv")
taxonomy_df.head()
```

```python
# Create human-readable description for each taxonomy code
def build_taxonomy_description(row):
    if pd.notna(row["Specialization"]):
        return f"{row['Classification']} - {row['Specialization']}"
    else:
        return row["Classification"]
```

```python
taxonomy_df["taxonomy_description"] = taxonomy_df.apply(build_taxonomy_description, axis=1)
```

```python
taxonomy_df.head()
```

```python
# Create the mapping dictionary: {code: description}
taxonomy_map = dict(zip(taxonomy_df["Code"], taxonomy_df["taxonomy_description"]))
```

```python
# taxonomy_map
```

```python
# Convert stringified list to list (if needed)
renamed_df["taxonomy_codes_list"] = renamed_df["taxonomy_codes_list"].apply(
    lambda x: eval(x) if isinstance(x, str) else x
)
```

```python
renamed_df.head()
```

```python
# Map taxonomy_codes_list to description list
def map_taxonomy_list(codes):
    return [taxonomy_map.get(code, code) for code in codes]
```

```python
renamed_df["taxonomy_descriptions_list"] = renamed_df["taxonomy_codes_list"].apply(map_taxonomy_list)
```

```python
renamed_df.head()
```

```python
# Map primary_taxonomy_code to description
renamed_df["primary_taxonomy_description"] = renamed_df["primary_taxonomy_code"].map(taxonomy_map)
```
